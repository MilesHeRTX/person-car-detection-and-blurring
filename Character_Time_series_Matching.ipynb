```python
import cv2
import os

def video_to_frames(video_path, output_dir):
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Open the video file
    cap = cv2.VideoCapture(video_path)

    frame_count = 0
    while True:
        ret, frame = cap.read()

        if not ret:
            break  # End of video

        # Rotate the frame by 180 degrees
        rotated_frame = cv2.rotate(frame, cv2.ROTATE_180)

        # Save the rotated frame
        output_path = os.path.join(output_dir, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(output_path, rotated_frame)

        frame_count += 1

    cap.release()
    print(f"Extracted and rotated {frame_count} frames to {output_dir}")

if __name__ == "__main__":
    video_path = "/content/gdrive/MyDrive/Copy of GX010077.MP4"  # Replace with your video file path
    output_dir = "/content/Character-Time-series-Matching/Vietnamese/Vietnamese_imgs"  # Replace with your desired output directory
    video_to_frames(video_path, output_dir)
```

```python
from PIL import Image, ImageFilter
import cv2
import numpy as np

def blur_people_in_original(image_path, model):
    # Load the original image and get its dimensions
    original_image = Image.open(image_path)
    original_width, original_height = original_image.size

    # Resize the image to 1280 for detection
    resized_image = original_image.resize((1280, int(1280 * original_height / original_width)))
    resized_width, resized_height = resized_image.size

    # Convert resized image to a format compatible with the model
    resized_image_array = np.array(resized_image)

    # Detect people in the resized image (using your model)
    # Assuming `model.detect()` returns bounding boxes as (x, y, width, height)
    detections = model.detect(resized_image_array)

    # Calculate scaling factors between original and resized image
    scale_x = original_width / resized_width
    scale_y = original_height / resized_height

    # Loop through each detected bounding box
    for box in detections:
        # Get box coordinates and resize them back to the original image's scale
        x, y, w, h = box
        x_orig, y_orig = int(x * scale_x), int(y * scale_y)
        w_orig, h_orig = int(w * scale_x), int(h * scale_y)

        # Crop the person area from the original image
        person_area = original_image.crop((x_orig, y_orig, x_orig + w_orig, y_orig + h_orig))

        # Apply blur to the cropped area
        blurred_area = person_area.filter(ImageFilter.GaussianBlur(15))

        # Paste the blurred area back into the original image
        original_image.paste(blurred_area, (x_orig, y_orig))

    return original_image

# Usage
# Assuming `model` is your pre-loaded detection model
blurred_image = blur_people_in_original("path_to_image.jpg", model)
blurred_image.show()  # Or save the result with blurred_image.save("output.jpg")
```
